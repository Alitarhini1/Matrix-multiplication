```python
import numpy as np
def multiply_matrices(matrix1, matrix2):
    result=np.dot(matrix1, matrix2)
    return result
matrix_a=np.random.rand(3,3)
matrix_b=np.random.rand(3,3)
print("Matrix A:")
print(matrix_a)
print("\n Matrix B:")
print(matrix_b)
result_matrix=multiply_matrices(matrix_a,matrix_b)
print("\n Result Matrix:")
print(result_matrix)
```
